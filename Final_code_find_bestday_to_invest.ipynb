```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("mf_data.csv")
df.columns = ['date','price']
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
def find_data (data, best_day=3):
    
    years = data['date'].dt.year.unique()
    
    final_df = pd.DataFrame()
    
    for year in years:
        year_df = data[data['date'].dt.year == year]
        
        months = year_df['date'].dt.month.unique()
        
        for month in months:
            month_df = year_df[year_df['date'].dt.month == month]
            
            month_df = month_df.sort_values(by='date')
            
            selected_data = month_df[month_df['date'].dt.day >= best_day].head(1)
            
            final_df = pd.concat([final_df, selected_data], axis=0)
    
    final_df = final_df.reset_index(drop=True)
    return final_df
```

```python
# find the total expence investment each day.

def make_dataframe ():
    
    last_df = pd.DataFrame(columns=['invest_day', 'NO_investment', 'Sum_NAV'])
    
    for day in range(1,27): 
        data_df = find_data(df, best_day = day)
        sum_of_nav = data_df['price'].sum()
        number_of_investment = len(data_df)
        
        # last_df = last_df.append({'invest_day' : day, 'Sum_NAV' : sum_of_nav}, ignore_index=True)
        temp_df = pd.DataFrame({'invest_day' : [day], 'NO_investment' : [number_of_investment], 'Sum_NAV' : [sum_of_nav]})
        last_df = pd.concat([last_df, temp_df], axis=0)

        
    last_df = last_df.reset_index(drop=True)
    
    return last_df
```

```python
# data best day to worst day

final = make_dataframe ()
sorted_final = final.sort_values(by='Sum_NAV')
sorted_final.reset_index(drop=True)
```

```python
# Best and worst day to Invest in month

final = make_dataframe ()

min_value = final['Sum_NAV'].min()
max_value = final['Sum_NAV'].max()

min_row = final[final['Sum_NAV'] == min_value]
max_row = final[final['Sum_NAV'] == max_value]

best_and_worst_df = pd.concat([min_row, max_row], axis=0)
best_and_worst_df['Name']=['Best Day','Worsr Day']
best_and_worst_df.set_index('Name',inplace=True)

best_and_worst_df
```
